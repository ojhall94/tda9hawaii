```python
import lightkurve as lk
import warnings
warnings.filterwarnings('ignore')
```

```python
target = 'HIP 116158'

datalist = lk.search_lightcurvefile(target, cadence='short')
datalist
```

```python
lc = datalist.download_all().stitch()
```

```python
lc0 = lc.normalize().remove_outliers().remove_nans()
lc0.plot();
```

```python
lc1 = lc.normalize(unit='ppm').remove_outliers().remove_nans()
lc1.plot();
```

```python
lc3 = lc.normalize(unit='ppm').remove_outliers().remove_nans() - 1e6
lc3.plot(); 
```

```python
lc0.to_periodogram(normalization = 'psd').plot()
```

```python
lc1.to_periodogram(normalization = 'psd', minimum_frequency=380, maximum_frequency=500).plot()
```

```python
lc3.to_periodogram(normalization='psd').plot()
```
